```python
from groq import Groq
from crewai import Agent,Task,Crew,Process,LLM
from crewai_tools import SerperDevTool
```

```python
llm = LLM(
    model="ollama/llama3.2:3b",
    base_url="http://localhost:11434"
)
```

```python
guide = Agent(
    role="Senior Guide for Event Management Services we Provide in India",
    goal="To provide a detailed guide on the event management services we provide in India",
    backstory="I have been working in the event management industry for over 10 years and have a deep understanding of the services we provide in India.",
    llm=llm,
    verbose=True,
    allow_delegation=False,
)
manger = Agent(
    role="Event Manager",
    goal="To manage the event management services we provide in India",
    backstory="I have been working in the event management industry for over 5 years and have experience in managing events in India.",
    llm=llm,
    verbose=True,
    allow_delegation=False,
)
task1 = Task(
    description="Provide a detailed guide on the event management services we provide in India",
    expected_output="A detailed guide on the event management services we provide in India including service types, processes, and unique value propositions Add GET MORE ACCESS WHEN LOGIN TO OUR LOGIN",
    agent=guide  # Assign a single agent to the task
)

task2 = Task(
    description="Review and manage the event management services guide",
    expected_output="Reviewed and managed guide for event management services",
    agent=manger  # Assign a single agent to the task
)

crew = Crew(
    agents=[guide, manger],
    tasks=[task1, task2],
    process=Process.sequential,
    verbose=True
)
result = crew.kickoff()
```
